```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
#reading image
img = cv2.imread('/Users\laksh\OneDrive\Pictures\FB_IMG_1483023883211.jpg')
```

```python
#display image
plt.imshow(img)
```

```python
#displaying image on another window and exit when any key is pressed.
cv2.imshow('meme',img)
cv2.waitKey(0)

cv2.destroyAllWindows()
```

```python
#converting from RGB to Gray scale
gray_img= cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2.imshow('gray_meme',gray_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python

plt.imshow(cv2.cvtColor(gray_img, cv2.COLOR_GRAY2RGB))
```

```python
#writing this image to disk
meme= cv2.imwrite('meme.png',gray_img)
```

```python
print(meme)
```

```python
#converting RGB to Binary(using binary thresholding)
#1. Convert RGB to gray thats the 1st argument of threshold function.
#2. The second argument is the threshold, ie, below threshold the image will have same value ie 127 and above the threshold the image will have same value ie 255 which is our third argument.
#3. Next argument is the type of thresholding.
ret, binary_image=cv2.threshold(gray_img,127,255,cv2.THRESH_BINARY)
cv2.imshow('binary image',binary_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.imshow(cv2.cvtColor(binary_image,cv2.COLOR_GRAY2RGB))
```

```python
#simple inverse binary thresholding

ret, binary_image_inv = cv2.threshold(gray_img, 127, 255, cv2.THRESH_BINARY_INV) 
plt.imshow(cv2.cvtColor(binary_image_inv, cv2.COLOR_GRAY2RGB))
```

```python
#adaptive thresholding using cv2.ADAPTIVE_THRESH_MEAN_C
advth = cv2.adaptiveThreshold(gray_img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,29,1)
plt.imshow(cv2.cvtColor(advth, cv2.COLOR_GRAY2RGB))
```

```python
#Adaptive thresholding using cv2.ADAPTIVE_THRESH_GAUSSIAN_C
advth2= cv2.adaptiveThreshold(gray_img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,29,1)
plt.imshow(cv2.cvtColor(advth2, cv2.COLOR_GRAY2RGB))
```

```python
#Otsu's thresholding
ret2,th2 = cv2.threshold(gray_img,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
plt.imshow(cv2.cvtColor(th2,cv2.COLOR_GRAY2RGB))
```
